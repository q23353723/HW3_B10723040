```python
import pandas as pd
import sklearn
import numpy as np
from sklearn import preprocessing
```

```python
coke = pd.read_csv("./data/Coke.csv")[['Coke','rank']].to_numpy()
coke[:, 0] = preprocessing.scale(coke[:, 0], axis=0, with_mean=True, with_std=True, copy=True)
coke_label = pd.read_csv("./data/Coke.csv")['label'].to_numpy()
pepsi = pd.read_csv("./data/Pepsi.csv")[['Pepsi', 'rank']].to_numpy()
pepsi[:, 0] = preprocessing.scale(pepsi[:, 0], axis=0, with_mean=True, with_std=True, copy=True)
pepsi_label = pd.read_csv("./data/Pepsi.csv")['label'].to_numpy()
up = pd.read_csv("./data/7Up.csv")[['7Up', 'rank']].to_numpy()
up[:, 0] = preprocessing.scale(up[:, 0], axis=0, with_mean=True, with_std=True, copy=True)
up_label = pd.read_csv("./data/7Up.csv")['label'].to_numpy()
sprite = pd.read_csv("./data/Sprite.csv")[['Sprite', 'rank']].to_numpy()
sprite[:, 0] = preprocessing.scale(sprite[:, 0], axis=0, with_mean=True, with_std=True, copy=True)
sprite_label = pd.read_csv("./data/Sprite.csv")['label'].to_numpy()
latte = pd.read_csv("./data/Latte.csv")[['Latte', 'rank']].to_numpy()
latte[:, 0] = preprocessing.scale(latte[:, 0], axis=0, with_mean=True, with_std=True, copy=True)
latte_label = pd.read_csv("./data/Latte.csv")['label'].to_numpy()
espresso = pd.read_csv("./data/Espresso.csv")[['Espresso', 'rank']].to_numpy()
espresso[:, 0] = preprocessing.scale(espresso[:, 0], axis=0, with_mean=True, with_std=True, copy=True)
espresso_label = pd.read_csv("./data/Espresso.csv")['label'].to_numpy()
cappuccino = pd.read_csv("./data/Cappuccino.csv")[['Cappuccino', 'rank']].to_numpy()
cappuccino[:, 0] = preprocessing.scale(cappuccino[:, 0], axis=0, with_mean=True, with_std=True, copy=True)
cappuccino_label = pd.read_csv("./data/Cappuccino.csv")['label'].to_numpy()
```

```python
datapoints = np.concatenate((coke, pepsi, up, sprite, latte, espresso, cappuccino))
datapoints_label = np.concatenate((coke_label, pepsi_label, up_label, sprite_label, latte_label, espresso_label, cappuccino_label))
print(datapoints.shape)
print(datapoints_label.shape)
```

```python
color_map = []
for label in datapoints_label:
    if label == 'A':
        color_map.append("red")
    elif label == 'B':
        color_map.append("blue")
    elif label == 'C':
        color_map.append("yellow")
    elif label == 'D':
        color_map.append("pink")
    elif label == 'E':
        color_map.append("orange")
    elif label == 'F':
        color_map.append("black")
    elif label == 'G':
        color_map.append("green")
```

```python
from sklearn import manifold
```

```python
X_tsne = manifold.TSNE(n_components=2, init='random', random_state=5, verbose=1).fit_transform(datapoints)
```

```python
df = pd.DataFrame(dict(Feature_1=X_tsne[:,0], Feature_2=X_tsne[:,1], label=color_map))
df.plot(x="Feature_1", y="Feature_2", kind="scatter", c='label', colormap='viridis')
```

```python
import dash
from dash.dependencies import Input, Output
import dash_core_components as dcc
import dash_html_components as html
import json

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

result_dataset = []
for i in range(1000):
    if i < 200:
        result_dataset.append({
            'x': [X_tsne[i,0]],
            'y': [X_tsne[i,1]],
            'text': ['Coke'],
            'name': 'Coke',
            'mode': 'markers',
            'marker': {'size': 12, 'color': 'red'}
        })
    elif i < 300:
        result_dataset.append({
            'x': [X_tsne[i,0]],
            'y': [X_tsne[i,1]],
            'text': ['Pepsi'],
            'name': 'Pepsi',
            'mode': 'markers',
            'marker': {'size': 12, 'color': 'blue'}
        })
    elif i < 400:
        result_dataset.append({
            'x': [X_tsne[i,0]],
            'y': [X_tsne[i,1]],
            'text': ['7Up'],
            'name': '7Up',
            'mode': 'markers',
            'marker': {'size': 12, 'color': 'yellow'}
        })
    elif i < 600:
        result_dataset.append({
            'x': [X_tsne[i,0]],
            'y': [X_tsne[i,1]],
            'text': ['Sprite'],
            'name': 'Sprite',
            'mode': 'markers',
            'marker': {'size': 12, 'color': 'pink'}
        })
    elif i < 700:
        result_dataset.append({
            'x': [X_tsne[i,0]],
            'y': [X_tsne[i,1]],
            'text': ['Latte'],
            'name': 'Latte',
            'mode': 'markers',
            'marker': {'size': 12, 'color': 'orange'}
        })
    elif i < 800:
        result_dataset.append({
            'x': [X_tsne[i,0]],
            'y': [X_tsne[i,1]],
            'text': ['Espresso'],
            'name': 'Espresso',
            'mode': 'markers',
            'marker': {'size': 12, 'color': 'black'}
        })
    else:
        result_dataset.append({
            'x': [X_tsne[i,0]],
            'y': [X_tsne[i,1]],
            'text': ['Cappuccino'],
            'name': 'Cappuccino',
            'mode': 'markers',
            'marker': {'size': 12, 'color': 'green'}
        })

app.layout = html.Div([
    dcc.Graph(
        id='basic-interactions',
        figure={
            'data': result_dataset
        }
    ),

    html.Div([
        dcc.Markdown("""
            **Hover Data**

            Mouse over values in the graph.
        """.replace('   ', '')),
        html.Pre(id='hover-data')
    ]),

    html.Div([
        dcc.Markdown("""
            **Click Data**

            Click on points in the graph.
        """.replace('    ', '')),
        html.Pre(id='click-data'),
    ]),

    html.Div([
        dcc.Markdown("""
            **Selection Data**

            Choose the lasso or rectangle tool in the graph's menu
            bar and then select points in the graph.
        """.replace('    ', '')),
        html.Pre(id='selected-data'),
    ])
])


@app.callback(
    Output('hover-data', 'children'),
    [Input('basic-interactions', 'hoverData')])
def display_hover_data(hoverData):
    #
    # This is where you can access the hover data
    # This function will get called automatically when you hover over points
    # hoverData will be equal to an object with that data
    # You can compute something off of this data, and return it to the front-end UI
    # 


    return json.dumps(hoverData, indent=2)


@app.callback(
    Output('click-data', 'children'),
    [Input('basic-interactions', 'clickData')])
def display_click_data(clickData):
    # Similarly for data when you click on a point
    return json.dumps(clickData, indent=2)


@app.callback(
    Output('selected-data', 'children'),
    [Input('basic-interactions', 'selectedData')])
def display_selected_data(selectedData):
    # Similarly for data when you select a region
    return json.dumps(selectedData, indent=2)


if __name__ == '__main__':
    app.run_server()
```
